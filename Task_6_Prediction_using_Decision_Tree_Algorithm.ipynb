```python
# Import libraries 
import sklearn.datasets as datasets
import numpy as np
import pandas as pd
```

```python
# Load data set 
iris=datasets.load_iris()
```

```python
iris.feature_names
```

```python
iris.target
```

```python
iris.target_names
```

### Creating the Dataframe

```python
data=pd.DataFrame(np.c_[iris.data,iris.target],columns=[list(iris.feature_names)+['target']])
data
```

```python
data.shape
```

```python
data.info()
```

```python
data.describe()
```

```python
data.isnull().sum()
```

### Split Data

```python
X= data.iloc[:,0:-1]
y=data.iloc[:,-1]
```

```python
from sklearn.model_selection import train_test_split
X_train,X_test,y_train,y_test=train_test_split(X,y,test_size=0.5,random_state=100)
print('Shape of X_train:',X_train.shape)
print('Shape of X_test:',X_test.shape)
print('Shape of y_train:',y_train.shape)
print('Shape of y_test:',y_test.shape)
```

### Train decision tree classifier

```python
from sklearn.tree import DecisionTreeClassifier
```

```python
model=DecisionTreeClassifier(criterion='gini')
```

```python
model.fit(X_train,y_train)
```

```python
model.score(X_test,y_test)
```

```python
y_predict=model.predict(X_test)
y_predict
```

### Visualizing the Decision Tree

```python
from sklearn import tree
import matplotlib.pyplot as plt
from sklearn.metrics import accuracy_score,confusion_matrix
import seaborn as sns
```

```python
# Accuracy For training data
predict = model.predict(X_train)
print("Accuracy of training data : ",accuracy_score(predict,y_train)*100,"%")
print("Confusin matrix of training data :'\n' ",confusion_matrix(predict,y_train))
sns.heatmap(confusion_matrix(predict,y_train),annot = True,cmap = 'BuGn')
```

```python
feature=['sepal length (cm)','sepal width (cm)','petal length (cm)','petal width (cm)']
cl=['setosa','versicolor','virginica']

fig, axes = plt.subplots(nrows = 1, ncols = 1, figsize = (6,4), dpi = 400)

tree.plot_tree(model, feature_names = feature, class_names = cl, filled = True);
```
